```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
sns.kdeplot(data=df, x="size", shade=True)
plt.title("a KDE plot for the size feature of the tips dataset")
plt.show()
```

Observation: The kde plot shows the density and sizes of the tips, with the peak of curv being above 2 showing that the sizes of the tips are mostly around 2 and 7 having little to no tips at all

```python
plot = sns.boxplot(data=df, x="sex", y="total_bill" )
plt.title('The distribution of total bill by gender for the tips dataset.')
plt.show()
```

 The box plot above shows the total distribution of the bill by gender, indicating that the male has a higher bill compared to the female.

```python
sns.boxplot(data=df, x="day", y="tip")
plt.title("distribution of tips by days" )
plt.show()
```

```python
sns.boxplot(data=df, x="day", y="total_bill", hue="smoker")
plt.title("Distribution of Total Bill by Days considering smoker status of the customer")
plt.show()
```

The boxplot above both smokers and non smoker have the same amount of bill of thursady, and non smokers have a slightly higher bill on fridays. However, the plot indicates that during the weekend especially on sundays, smokers have a higher total bill than non smokers.

```python
sns.violinplot(data=df, x ="total_bill" , y = "time", hue="sex")
plt.title("Distribution of Total Bill by time given the genders")
plt.show()
```

```python
sns.violinplot(data=df, x="tip", y="day", hue="time");
plt.title("Distribution of tips by days considering time given the genders") 
plt.show()
```

```python
corr = df.corr()
ones = np.ones_like(corr, dtype='bool')
mask = np.triu(ones)
mask = mask[1:, :-1]
corr = corr.iloc[1:,:-1].copy()
fig, ax = plt.subplots(figsize=(8,6))
sns.heatmap(corr, mask=mask, annot=True, fmt=".1f", cmap='Greens', linewidths=.9, cbar_kws={'shrink':.99})
ax.xaxis.tick_bottom()
plt.title("correlations heatmap for the tips dataset. \n".upper(), fontdict={'fontsize': 18}, loc='left')
plt.show()
```
