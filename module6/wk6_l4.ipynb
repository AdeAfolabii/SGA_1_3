```python
df_customers = pd.DataFrame(customers)
df_customers
```

```python
#An inner merge operation between the orders table and products table
df_merge1 = pd.merge(df_order, df_products, how="inner")
df_merge1
```

```python
#. Using the resulting table to perform another inner merge operation with the customers table.
df_merge2 = pd.merge(df_merge1, df_customers, how="inner")
df_merge2
```

OBSERVATION: the inital operation merged the first 2 datasets and all elements of the dataset into one dataset, while the second operation merged the result of the first merge with the customers dataset.

```python
#Performing an inner merge operation between customers and orders tables while specifying the columns

df_merge3 = df_customers.merge(df_order, on="customer_id", how="left", suffixes=('_left', '_right'))
df_merge3
```

OBSERVATION: this operation merged the two datasets and returned column exisiting in thr two datasets.

```python
# Performing an outer merge between customers and products tables.
df_merge4 = df_customers.merge(df_products, left_index=True, right_index=True, how="outer", indicator=True)
df_merge4
```

ONBSERVATION: This merge returned all elements in both datasets. however, due to a dataset having more columns/values, it printed out NaN which is a null value indicating data points with values in the second dataset.

```python
#Perform a left merge between orders and customers tables

merge5 = df_order.merge(df_customers, how="left")
merge5
```

OBSERVATION: This merge returned all elements in both datasets. however, due to a dataset having more columns/values, it printed out NaN which is a null value indicating data points with values in the second dataset.

```python
# Perform a right merge between product and customers tables
merge6 = df_products.merge(df_customers, left_index=True, right_index=True, how="right", indicator=True)
merge6
```

```python
#Using the pandas concatenation method to join the three datasets together
df_all_merge = pd.concat([df_order, df_customers,df_products], axis=1)
df_all_merge
```
