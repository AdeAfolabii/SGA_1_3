```python
import matplotlib 
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# read csv file
df= pd.read_csv('anscombe_quartet_data.csv')
# check
df
```

```python
#FIRST PLOT
plt.figure(figsize=(10, 10))
x = [0,25]
y = [0,20]
x1 = df['x1']
y1 = df['y1'] 
plt.subplot(2, 2, 1)
plt.title('(x1,y1)')
plt.xlabel('x1')
plt.ylabel('y1')
plt.scatter(x1, y1, color ="black")
plt.plot(x,y)

x = [0,25]
y = [0, 20]
x2 = df['x2']
y2 = df['y2'] 
plt.subplot(2, 2, 2)
plt.title('(x2,y2)')
plt.xlabel('x2')
plt.ylabel('y2')
plt.scatter(x2, y2, color ="orange")
plt.plot(x,y)

x = [0,20]
y = [0, 25]
x3 = df['x3']
y3 = df['y3'] 
plt.subplot(2, 2, 3)
plt.title('(x3,y3)')
plt.xlabel('x3')
plt.ylabel('y3')
plt.scatter(x3, y3, color ="pink")
plt.plot(x,y)

x = [0,20]
y = [0,25]
x4 = df['x4']
y4 = df['y4'] 
plt.subplot(2, 2, 4)
plt.title('(x4,y4)')
plt.xlabel('x4')
plt.ylabel('y4')
plt.scatter(x4, y4, color ="red")
plt.plot([4, 14], [4, 12])

plt.subplots_adjust(wspace=0.6, hspace =0.5, bottom =0.1, top =0.9)
plt.show()
```

```python
#no2
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
x = range(0, 15)
y = range(0, 15)
x1 = df["x1"]
y1 = df["y1"]
ax1.scatter(x1, y1, color="gold")
ax1.plot(x, y)
ax1.set_title('x1y1')

x = range(0, 15)
y = range(0, 15)
x2 = df["x2"]
y2 = df["y2"]
ax2.scatter(x2, y2, color="pink")
ax2.plot(x, y)
ax2.set_title('x2y2')

x = range(0, 15)
y = range(0, 15)
x3 = df["x3"]
y3 = df["y3"]
ax3.scatter(x3, y3, color="purple")
ax3.plot(x, y)
ax3.set_title('x3y3')

x = range(0, 15)
y = range(0, 15)
x4 = df["x4"]
y4 = df["y4"]
ax4.scatter(x4, y4, color="yellow")
ax4.plot(x, y)
ax4.set_title('x4y4')

plt.subplots_adjust(wspace=0.8, hspace=0.6, bottom=0.1, top=0.9)
```
