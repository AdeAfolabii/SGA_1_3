```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
```

```python
#read csv file
df = pd.read_csv('company_sales_data.csv')
df.head()
```

```python
plt.plot(df["month_number"], df["facecream"], color="blue", ls="--", marker="o")
plt.plot(df["month_number"], df["facewash"], color="gold", ls=":", marker=">")
plt.plot(df["month_number"], df["toothpaste"], color="orange", ls="-.", marker="*")
plt.plot(df["month_number"], df["bathingsoap"], color="red", ls="-", marker="8")
plt.plot(df["month_number"], df["shampoo"], color="grey", marker="H")
plt.plot(df["month_number"], df["moisturizer"], color="purple")
plt.legend(["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo", "moisturizer"])
plt.xlabel("Month of the year")
plt.ylabel("Sales Data")
plt.title("Company Sales Data")
plt.show()
```
